```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
#sns.set(style="whitegrid", color_codes=True)
```

```python
df1 = pd.read_csv('bo.txt', index_col=None, header=0, sep='\t')
df2 = pd.read_csv('cc.txt', index_col=None, header=0, sep='\t')
df3 = pd.read_csv('mf.txt', index_col=None, header=0, sep='\t')
df4 = pd.read_csv('pc.txt', index_col=None, header=0, sep='\t')
```

```python
f, axes = plt.subplots(4, 1, figsize=(5, 15))
sns.barplot(x="transcripts", y="description", data=df1, orient="h", palette="Greys_r", ax=axes[0])
sns.barplot(x="transcripts", y="description", data=df2, orient="h", palette="Greys_r", ax=axes[1])
sns.barplot(x="transcripts", y="description", data=df3, orient="h", palette="Greys_r", ax=axes[2])
sns.barplot(x="transcripts", y="description", data=df4, orient="h", palette="Greys_r", ax=axes[3])
f.savefig("test.svg")
```
