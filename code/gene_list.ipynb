## Extract gene names from matrix

Python script for creating a list of up- or down-regulated genes from edgeR results

```python
import pandas as pd
```

```python
df = pd.read_csv('32_diffex.txt', index_col=1, header=0, sep=' ')
```

```python
df.head()
```

```python
# Sort values by index
df.sort_index(axis=0, ascending=True, inplace=False).head()
```

```python
# Sort values by column
df.sort_values('logFC', axis=0, ascending=True, inplace=False).head()
```

```python
df1 = df[['logFC', 'logCPM', 'PValue', 'FDR']]
```

```python
df1.head()
```

```python
df1.index.name = 'gene'
```

```python
df1.head()
```

```python
#df1.reset_index(inplace=True)
```

```python
for i, row in df1.iterrows():
    if row['logFC'] > 0:
        print(row.name, row['logFC'])
```

```python
for i, row in df1.iterrows():
    if row['logFC'] < 0:
        print(row.name, row['logFC'])
```
